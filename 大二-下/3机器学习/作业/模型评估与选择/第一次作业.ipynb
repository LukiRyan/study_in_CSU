```python
#性能度量：某池塘有500条鲤鱼，400只虾，100只鳖。现在以捕鲤鱼为目的撒网，逮着了300条鲤鱼，190只虾，10只鳖。请计算以下指标：查准率、查全率、F1，F0.5，F2，试用matlab/python画出Fa的变化趋势图（注a的取值范围为（0.3，2），且a的取值不少于15个点）。
```

```python
#查准率 = TP/(TP+FP)
#查全率 = TP/(TP+FN)
#F1 = 2*查准率*查全率/(查准率+查全率)
#F0.5 = 1.25*查准率*查全率/(0.25*查准率+查全率)
#F2 = 5*查准率*查全率/(4*查准率+查全率)
```

```python
fish = 500
shrimp = 400
turtle = 100
fish_catch = 300
shrimp_catch = 190
turtle_catch = 10
P = fish_catch/(fish_catch+shrimp_catch+turtle_catch)  #查准率
R = fish_catch/fish  #查全率
F1 = 2*P*R/(P+R)
F05 = 1.25*P*R/(0.25*P+R)
F2 = 5*P*R/(4*P+R)
print('查准率：',P)
print('查全率：',R)
print('F1：',F1)
print('F0.5：',F05)
print('F2：',F2)
```
